```python
taxis_jan19s=pd.read_csv("taxis_jan19s.csv")
taxis_feb19s=pd.read_csv("taxis_feb19s.csv")
taxis_mar19s=pd.read_csv("taxis_mar19s.csv")
taxis_apr19s=pd.read_csv("taxis_apr19s.csv")
taxis_may19s=pd.read_csv("taxis_may19s.csv")
taxis_jun19s=pd.read_csv("taxis_jun19s.csv")
taxis_jul19s=pd.read_csv("taxis_jul19s.csv")
taxis_aug19s=pd.read_csv("taxis_aug19s.csv")
```

```python
taxis_jantoaug19s=pd.concat([taxis_jan19s,taxis_feb19s,taxis_mar19s,taxis_apr19s,taxis_may19s,taxis_jun19s,taxis_jul19s,taxis_aug19s])
```

```python
taxis_jantoaug19s.head()
```

```python
taxis_jantoaug19s.columns
```

```python
taxis_jantoaug19s=taxis_jantoaug19s.reset_index(drop=True)
```

```python
taxis_jantoaug19s.info()
```

```python
# drop columns named "index" and "Unnamed: 0"
taxis_jantoaug19s=taxis_jantoaug19s.drop(columns=['index','Unnamed: 0'])
```

```python
taxis_jantoaug19s.columns
```

```python
taxis_jantoaug19s[taxis_jantoaug19s['congestion_surcharge'].isnull()]
```

```python
# fill 9752 NaN entries for congestion surcharge with 0
taxis_jantoaug19s['congestion_surcharge']=taxis_jantoaug19s['congestion_surcharge'].fillna(0.0)
```

```python
# rename tpep_dropoff_datetime and tpep_pickup_datetime with dropoff_datetime and pickup_datetime respectively
taxis_jantoaug19s=taxis_jantoaug19s.rename(columns={"tpep_dropoff_datetime": "dropoff_datetime", "tpep_pickup_datetime": "pickup_datetime"})
taxis_jantoaug19s.head()
```

```python
# create a new trip_duration_mins variable by finding the difference between pickup_datetime and dropoff_datetime
taxis_jantoaug19s["pickup_datetime"]=pd.to_datetime(taxis_jantoaug19s['pickup_datetime'])
taxis_jantoaug19s["dropoff_datetime"]=pd.to_datetime(taxis_jantoaug19s['dropoff_datetime'])
taxis_jantoaug19s['trip_duration_mins'] = taxis_jantoaug19s['dropoff_datetime'] - taxis_jantoaug19s['pickup_datetime']
taxis_jantoaug19s['trip_duration_mins']=taxis_jantoaug19s['trip_duration_mins']/np.timedelta64(1,'m')
taxis_jantoaug19s['trip_duration_mins']=np.around(taxis_jantoaug19s['trip_duration_mins'],decimals=2)
```

```python
# create a new pickup_dayofweek variable
taxis_jantoaug19s["pickup_dayofweek"]=pd.to_datetime(taxis_jantoaug19s['pickup_datetime']).dt.dayofweek
```

```python
# check for non-2019 observations
taxis_jantoaug19s.loc[(pd.to_datetime(taxis_jantoaug19s['pickup_datetime']).dt.year!=2019)|(pd.to_datetime(taxis_jantoaug19s['pickup_datetime']).dt.year!=2019)]
```

```python
# remove the 1 non-2019 observation
taxis_jantoaug19s.drop([46829], inplace=True)
```

```python
taxis_jantoaug19s['congestion_surcharge'].value_counts()
```

```python
# remove the 111 observations with negative congestion_surcharge
taxis_jantoaug19s = taxis_jantoaug19s[taxis_jantoaug19s["congestion_surcharge"]>=0]
```

```python
taxis_jantoaug19s[(taxis_jantoaug19s["fare_amount"]<=0)|(taxis_jantoaug19s["trip_distance"]<=0)|(taxis_jantoaug19s["total_amount"]<=0)|(taxis_jantoaug19s["trip_duration_mins"]<=0)]
```

```python
# remove the 993 observations that fail the assumptions of a "valid" ride
# the assumptions are positive fare_amount, positive trip_distance, positive total_amount and positive trip_duration_mins
taxis_jantoaug19s=taxis_jantoaug19s[taxis_jantoaug19s["fare_amount"]>0]
taxis_jantoaug19s=taxis_jantoaug19s[taxis_jantoaug19s["trip_distance"]>0]
taxis_jantoaug19s=taxis_jantoaug19s[taxis_jantoaug19s["total_amount"]>0]
taxis_jantoaug19s=taxis_jantoaug19s[taxis_jantoaug19s["trip_duration_mins"]>0]
```

```python
taxis_jantoaug19s.info()
```

```python
#drop 628 null values
taxis_jantoaug19s=taxis_jantoaug19s.dropna()
```

```python
taxis_jantoaug19s.info()
```

```python
taxis_jantoaug19s[(taxis_jantoaug19s["payment_type"]==3.0) | (taxis_jantoaug19s["payment_type"]==4.0)]
```

```python
# remove 505 taxi rides with no charge or disputed taxi fares
taxis_jantoaug19s = taxis_jantoaug19s.query("(payment_type!=3.0) & (payment_type!=4.0)")
taxis_jantoaug19s = taxis_jantoaug19s.astype({"payment_type":"int32"})
```

```python
taxis_jantoaug19s.describe()
```

```python
# drop VendorID attribute
taxis_jantoaug19s.drop(columns=["VendorID"],inplace=True)
```

```python
taxis_jantoaug19s=taxis_jantoaug19s.reset_index(drop=True)
taxis_jantoaug19s.drop([110480], inplace=True)
```

```python
taxis_jantoaug19s.info()
```

```python
# save taxi dataset to taxis_jantoaug19s.csv
taxis_jantoaug19s.to_csv("taxis_jantoaug19s.csv", index=False)
```

# Merging with NYC Weather Dataset

```python
df_weather=pd.read_csv("nyc_weather_jantoaug19.csv")
df_weather.head()
```

```python
# merge taxi dataset with weather dataset
df_agg=taxis_jantoaug19s.merge(df_weather,how="left",left_on=pd.to_datetime(taxis_jantoaug19s["pickup_datetime"]).dt.date, right_on=pd.to_datetime(df_weather['date']).dt.date)
```

```python
df_agg.info()
```

```python
# drop key_0 and location variables 
df_agg.drop(['key_0'], axis=1, inplace=True)
df_agg.drop(['location'],axis=1, inplace=True)
```

```python
df_agg.columns
```

```python
df_agg[(df_agg["precipitation"]=="T")|(df_agg["newsnow"]=="T")|(df_agg["snowdepth"]=="T")]
```

```python
# replace trace precipitation, newsnow and snowdepth values with 0
df_agg.replace({"precipitation":{'T':0.0}}, inplace=True)
df_agg.replace({"newsnow":{'T':0.0}}, inplace=True)
df_agg.replace({"snowdepth":{'T':0.0}}, inplace=True)
```

```python
df_agg.dtypes
```

```python
# cast precipitation, newsnow and snowdepth to float
df_agg=df_agg.astype({"precipitation":"float64", "newsnow":"float64", "snowdepth": "float64"})
```

```python
df_agg.info()
```

```python
#save df_agg to nyc_taxis_weather_jantoaug19s.csv
df_agg.to_csv("nyc_taxis_weather_jantoaug19s.csv", index=False)
```
